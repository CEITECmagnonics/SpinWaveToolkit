```python
import SpinWaveToolkit as SWT
import numpy as np
import matplotlib.pyplot as plt
import warnings
```

```python
kxi = np.linspace(-100e6, 100e6, 200)
lxi = 2*np.pi/(kxi*1e-6)
```

```python
B_angle = 0
Anis1_angle = 90
Anis2_angle = -Anis1_angle
```

```python
CoFeB = SWT.Material(Ms = 1329e3, 
                     Aex = 15e-12, 
                     alpha = 40e-4, 
                     gamma=30*2*np.pi*1e9)
```

```python
SAF = SWT.DispersionCharacteristic(kxi = kxi, 
                                   theta = np.deg2rad(90), 
                                   phi = np.deg2rad(B_angle),  
                                   boundaryCond = 1,  
                                   Bext = 100e-3, 
                                   material = CoFeB, 
                                   Ku = 1.5e3,
                                   Jbl = -0.67e-3,
                                   Jbq = -0.28e-3,
                                   d = 10e-9,
                                   s=0.6e-9, 
                                   d2=10e-9,
                                   phiAnis1=np.deg2rad(Anis1_angle), 
                                   phiAnis2=np.deg2rad(Anis2_angle),
                                   phiInit1=np.deg2rad(-45), 
                                   phiInit2=np.deg2rad(45)
                                  )
```

```python
f = SAF.GetDispersionSAFMNumeric()/(1e9*2*np.pi) 
SAF1 = f[2,:] #acoustic mode
SAF2 = f[3,:] #optic mode
```

```python
phi1, phi2 = SAF.GetPhisSAFM()
print(np.rad2deg(phi1), np.rad2deg(phi2))
```

```python
# Function to plot angle with arrow
def plot_angle(x, y, angle, color, radius=1):
    ax = plt.gca()
    ax.plot([0, x], [0, y], 'k-')  # Plot line from origin to point
    ax.annotate('', xy=(x, y), xytext=(0, 0),
                arrowprops=dict(arrowstyle="->", color=color))  # Plot arrow
    ax.text(0.5*x, 0.5*y, "{:.1f}°".format(angle), ha='center', va='center')  # Label angle rounded to one decimal place
    #ax.text(0.5*x, 0.5*y, str(angle) + '°', ha='center', va='center')  # Label angle

# Sample angles
angles_anis = [Anis1_angle, Anis2_angle]   
angles_M = [np.rad2deg(phi1), np.rad2deg(phi2)]
angles_B = [B_angle]

# Plot angles
plt.figure(figsize=(3, 3))
for angle in angles_anis:
    radians = np.deg2rad(angle)
    x = np.cos(radians)
    y = np.sin(radians)
    plot_angle(x, y, angle, 'blue')

for angle in angles_M:
    radians = np.deg2rad(angle)
    x = np.cos(radians)
    y = np.sin(radians)
    plot_angle(x, y, angle, 'red')

for angle in angles_B:
    radians = np.deg2rad(angle)
    x = np.cos(radians)
    y = np.sin(radians)
    plot_angle(x, y, angle, 'green')

# Set limits and aspect ratio
plt.xlim(-1, 1)
plt.ylim(-1, 1)
plt.gca().set_aspect('equal', adjustable='box')

# Set labels and title
plt.title('B and ku vs canting angles\n B green ku blue M red')

# Show plot
plt.grid(True)
plt.show()
```

```python
from matplotlib.ticker import FuncFormatter
warnings.filterwarnings('ignore')

# Define the formula/function
def custom_formatter(x, pos):
    return "{:.0f}".format((2*np.pi/(x))*1000)

plt.figure()
plt.axhline(11.5, color='red', linestyle='--')
plt.axvline(-75, color='red', linestyle='--')
plt.axvline(0, color='gray', linestyle='--')
plt.plot(kxi*1e-6, SAF1);
plt.plot(kxi*1e-6, SAF2);
plt.xlabel('k-vector (rad/um)');
plt.ylabel('Frequency (GHz)');
#plt.legend(['acoustic'])
plt.title('Dispersion relation of SAF, CFB 10nm, B = 100 mT')

plt2 = plt.twiny()
plt2.spines['top'].set_visible(True)
#plt2.plot(kxi*1e-6,np.zeros(200))
plt.xlim(min(kxi*1e-6), max(kxi*1e-6))
plt.xlabel('Wavelength (nm)')
plt2.xaxis.set_major_formatter(FuncFormatter(custom_formatter))
plt.grid(True)
plt.show()
```
